# **1 - Leitura da base de dados**

```python
import pandas as pd

url = "https://raw.githubusercontent.com/sthemonica/alura-voz/main/Dados/Telco-Customer-Churn.json"
df = pd.read_json(url)
df.head()
```

# **3 - Organização do arquivo json**

```python
# Função para "desenrolar" o arquivo json
def unwrap(line,column):
    return line[column]
```

```python
# Definindo as colunas para serem abertas
columns = ['customer', 'phone','internet','account', 'Charges']

for c in columns:
    keys = df[c].iloc[0].keys()
    for key in keys:
        df[key] = df[c].apply(lambda x: unwrap(x,key))

df.drop(columns, inplace=True, axis = 1)
df.head()
```

# **5 - Limpeza dos dados**

## **5.1 - Limpando a coluna *Churn***

```python
# Identificando as linhas com valores nulos
ind = []
cont = 0
for line in df.Churn:
    if line == '' or line == ' ' or line == None:
        ind.append(cont)
    cont = cont + 1

# Removendo as linhas em branco
df.drop( ind, axis=0, inplace = True )
```

## **5.2 - Limpando a coluna Total**

```python
df = df.loc[df.Total != ' ']
df.head()
```

```python
# Tranformando a coluna Total em float
df["Total"] = df.Total.astype(float)
df.info()
```

## **5.3 - Excluindo a coluna de *customerID***

```python
# Avaliando a coluna tenure
df.drop('customerID', axis=1, inplace=True)
df.head()
```

## **5.4 - Adicionando um nova coluna de gastos diários**

```python
df.insert(18,'Daily',df.Monthly/30)
df.head()
```

# **6 - Tradução das colunas**

```python
trad = {'Churn':'Status_Cliente','gender':'Sexo','SeniorCitizen':'Aposentado','Partner':'Casado',
        'Dependents':'Dependentes', 'tenure':'Meses_Contrato', 'PhoneService':'Assinatura_Telefone', 'MultipleLines':'Assinatura_MultiTelefone',
        'InternetService':'Assinatura_Internet', 'OnlineSecurity':'Assinatura_Seguro_online', 'OnlineBackup':'Assinatura_Backup', 
        'DeviceProtection':'Assinatura_Seguro_dispositivo', 'TechSupport':'Assinatura_Suporte_tecnico', 'StreamingTV':'Assinatura_TV',
        'StreamingMovies':'Assinatura_Filmes','Contract':'Tipo_Contrato','PaperlessBilling':'Fatura_Online', 'PaymentMethod':'Forma_Pagamento',
        'Daily':'Valor_Diario','Monthly':'Valor_Mensal', 'Total':'Valor_Total'
        }
df.rename( columns=trad, inplace=True )
df.head()
```

# **7 - Padronização dos dados**

## **7.1 - Dataframe com respostas padronizada do Tipo Yes/No**

```python
df_obj = df.copy()
df_obj['Aposentado'] = df_obj['Aposentado'].map({0:'No', 1:'Yes'})
df_obj['Status_Cliente'] = df_obj['Status_Cliente'].map({'No':'Ativo', 'Yes':'Inativo'})
df_obj.head()
```

## **7.2 - Dataframe com respostas padrão do tipo 0/1**

```python
mappings = [] # mapeamento dos dados
df_num = df.copy()
for column in df.columns:
  if (df_num[column].dtypes != 'int64') & (df_num[column].dtypes != 'float64'):
    values = df_num[column].unique()
    mapping = dict()
    count = 0
    for value in values:
      mapping[value] = count
      count += 1 
    mappings.append([column, mapping])
    df_num[column] = df_num[column].map(mapping)
df_num
```

```python
# Dicionário do mapeamento
mappings
```

```python
# Exportando o DataFrame com padrões Yes/No
df_obj.to_csv('dados_tratados_obj.csv', index=False)

# Exportando o DataFrame com padrões 1/0
df_num.to_csv('dados_tratados_num.csv', index=False)
```
