# **1 - Leitura da base de dados**

```python
import pandas as pd

url = "https://raw.githubusercontent.com/sthemonica/alura-voz/main/Dados/Telco-Customer-Churn.json"
df = pd.read_json(url)
df.head()
```

# **3 - Organização do arquivo json**

```python
# Função para "desenrolar" o arquivo json
def unwrap(line,column):
    return line[column]
```

```python
# Definindo as colunas para serem abertas
columns = ['customer', 'phone','internet','account', 'Charges']

for c in columns:
    keys = df[c].iloc[0].keys()
    for key in keys:
        df[key] = df[c].apply(lambda x: unwrap(x,key))

df.drop(columns, inplace=True, axis = 1)
df.head()
```

# **5 - Limpeza dos dados**

## **5.1 - Limpando a coluna *Churn***

```python
# Identificando as linhas com valores nulos
ind = []
cont = 0
for line in df.Churn:
    if line == '' or line == ' ' or line == None:
        ind.append(cont)
    cont = cont + 1

# Removendo as linhas em branco
df.drop( ind, axis=0, inplace = True )
```

## **5.2 - Limpando a coluna Total**

```python
df = df.loc[df.Total != ' ']
df.head()
```

```python
# Tranformando a coluna Total em float
df["Total"] = df.Total.astype(float)
df.info()
```

```python
# Avaliando a coluna tenure
df.describe()
```

# **6 - Padronização dos dados**

## **6.1 - Dataframe com respostas padronizada do Tipo Yes/No**

```python
df['SeniorCitizen'] = df['SeniorCitizen'].map({0:'No', 1:'Yes'})
df_obj = df
df_obj.drop('customerID',axis=1,inplace=True)
df_obj.head()
```

## **6.2 - Dataframe com respostas padrão do tipo 0/1**

```python
mappings = []
# df.drop('customerID', axis=1, inplace=True)
for column in df.columns:
  if (df[column].dtypes != 'int64') & (df[column].dtypes != 'float64'):
    values = df[column].unique()
    mapping = dict()
    count = 0
    for value in values:
      mapping[value] = count
      count += 1 
    mappings.append([column, mapping])
    df[column] = df[column].map(mapping)
df
```
